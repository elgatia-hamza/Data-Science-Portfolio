## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
os.listdir()
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, -1].values
```

## Training the Random Forest Regression model on the dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10)
regressor.fit(X, y)
```

## Predicting the Test set results

```python
ypred = regressor.predict([[6.5]])
print(ypred)
```

## Visualising the Training set results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Salary vs Experience (Random Foreset results)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Training set results (with higher resolution)

```python
Xgrid = np.arange(min(X), max(X)+0.1, 0.1)
Xgrid=Xgrid.reshape(len(Xgrid),1)
plt.scatter(X, y, color = 'red')
plt.plot(Xgrid, regressor.predict(Xgrid), color = 'blue')
plt.title('Salary vs Experience (Random Foreset results)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
